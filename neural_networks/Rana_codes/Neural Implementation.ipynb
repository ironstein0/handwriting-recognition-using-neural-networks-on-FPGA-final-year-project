```python
import mnist_loader
training_data, validation_data, test_data = mnist_loader.load_data_wrapper()

import network
```

```python
# creating a network (loading a class)
net = network.Network([784, 30, 10])
```

```python
# start training with 30 epochs, mini_batch_size = 10 , eta = 3
epoch_efficiency_list = net.SGD(training_data,10,5,1.3,test_data = test_data)
```

```python
import matplotlib.pyplot as ppl
print(epoch_efficiency_list)
t = []
for i in range(len(epoch_efficiency_list)): t.append(i)
print(t)
ppl.plot(t,epoch_efficiency_list)
ppl.show()
```
